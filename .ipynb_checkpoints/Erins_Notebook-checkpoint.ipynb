```python
import json
import os

with open("countries/countries.json") as jsonfile:
    countries_json = json.load(jsonfile)
```

```python
common_name = []
official_name = []
country_code =[]
independent = []
status = []
capital = []
region = []
subregion = []
lat = []
long = []

for country in countries_json:
    common_name.append(country['name']['common'])
    official_name.append(country['name']['official'])
    country_code.append(country['cca3'])
    independent.append(country['independent'])
    status.append(country['status'])
    capital.append(country['capital'])
    region.append(country['region'])
    subregion.append(country['subregion'])
    lat.append(country['latlng'][0])
    long.append(country['latlng'][1])
```

```python
countriesDF = pd.DataFrame({"common_name":common_name, "official_name": official_name, "country_code": country_code, "independent":independent, "status": status, "capital":capital,"region":region,"subregion":subregion, "lat":lat, "long":long })
```
